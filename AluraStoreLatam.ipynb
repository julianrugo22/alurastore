### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

# 1. Análisis de facturación



```python
ingresotienda1 = tienda1['Precio'].sum()
ingresotienda2 = tienda2['Precio'].sum()
ingresotienda3 = tienda3['Precio'].sum()
ingresotienda4 = tienda4['Precio'].sum()
```

```python
plt.bar(['Tienda 1','Tienda 2','Tienda 3','Tienda 4'],[ingresotienda1,ingresotienda2,ingresotienda3,ingresotienda4], color = ['cornflowerblue','lightcoral','mediumpurple','mediumseagreen'])
plt.xlabel('Tiendas')
plt.ylabel('Ingreso')
plt.show()
```

# 2. Ventas por categoría

```python
ventasporcategoria1 = tienda1['Categoría del Producto'].value_counts().sort_index()
ventasporcategoria2 = tienda2['Categoría del Producto'].value_counts().sort_index()
ventasporcategoria3 = tienda3['Categoría del Producto'].value_counts().sort_index()
ventasporcategoria4 = tienda4['Categoría del Producto'].value_counts().sort_index()
```

```python
plt.figure(figsize=(17,2))
plt.bar(ventasporcategoria1.index.to_list(), ventasporcategoria1.to_list(), color = 'cornflowerblue', width = 0.1)
plt.bar(list(x + 0.1 for x in range(0,8)), ventasporcategoria2.to_list(), color = 'lightcoral', width = 0.1)
plt.bar(list(x + 0.2 for x in range(0,8)), ventasporcategoria3.to_list(), color = 'mediumpurple', width = 0.1)
plt.bar(list(x + 0.3 for x in range(0,8)), ventasporcategoria4.to_list(), color = 'mediumseagreen', width = 0.1)
plt.xlabel('Categoría')
plt.ylabel('Ventas')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
promediocalif1 = tienda1['Calificación'].mean()
promediocalif2 = tienda2['Calificación'].mean()
promediocalif3 = tienda3['Calificación'].mean()
promediocalif4 = tienda4['Calificación'].mean()

```

```python
plt.bar(['Tienda 1','Tienda 2','Tienda 3','Tienda 4'],[promediocalif1,promediocalif2,promediocalif3,promediocalif4], color = ['cornflowerblue','lightcoral','mediumpurple','mediumseagreen'])
plt.xlabel('Tiendas')
plt.ylabel('Calificación Promedio')
plt.ylim(3.5,5)
plt.show()
```

```python
calificaciones1 = tienda1['Calificación'].to_list()
calificaciones2 = tienda2['Calificación'].to_list()
calificaciones3 = tienda3['Calificación'].to_list()
calificaciones4 = tienda4['Calificación'].to_list()

plt.boxplot([calificaciones1,calificaciones2,calificaciones3,calificaciones4], labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
plt.show()
```

# 4. Productos más y menos vendidos

```python
masvendidos1 = tienda1['Producto'].value_counts()[0:3]
menosvendidos1 = tienda1['Producto'].value_counts()[-3:]

masvendidos2 = tienda2['Producto'].value_counts()[0:3]
menosvendidos2 = tienda2['Producto'].value_counts()[-3:]

masvendidos3 = tienda3['Producto'].value_counts()[0:3]
menosvendidos3 = tienda3['Producto'].value_counts()[-3:]

masvendidos4 = tienda4['Producto'].value_counts()[0:3]
menosvendidos4 = tienda4['Producto'].value_counts()[-3:]
```

```python
fig, axes = plt.subplots(4, 1, figsize=(16, 8), sharey=True)
plt.subplot(4,1,1)
plt.title('Tienda 1, más y menos vendidos')
plt.bar(masvendidos1.index.to_list(), masvendidos1.to_list(), color = 'mediumseagreen', width = 0.1,zorder = 3)
plt.bar(menosvendidos1.index.to_list(), menosvendidos1.to_list(), color = 'lightcoral', width = 0.1,zorder = 3)
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.grid(axis='y',zorder = 0)

plt.subplot(4,1,2)
plt.title('Tienda 2, más y menos vendidos')
plt.bar(masvendidos2.index.to_list(), masvendidos2.to_list(), color = 'mediumseagreen', width = 0.1,zorder = 3)
plt.bar(menosvendidos2.index.to_list(), menosvendidos2.to_list(), color = 'lightcoral', width = 0.1,zorder = 3)
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.grid(axis='y',zorder = 0)

plt.subplot(4,1,3)
plt.title('Tienda 3, más y menos vendidos')
plt.bar(masvendidos3.index.to_list(), masvendidos3.to_list(), color = 'mediumseagreen', width = 0.1,zorder = 3)
plt.bar(menosvendidos3.index.to_list(), menosvendidos3.to_list(), color = 'lightcoral', width = 0.1,zorder = 3)
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.grid(axis='y',zorder = 0)

plt.subplot(4,1,4)
plt.title('Tienda 4, más y menos vendidos')
plt.bar(masvendidos4.index.to_list(), masvendidos4.to_list(), color = 'mediumseagreen', width = 0.1,zorder = 3)
plt.bar(menosvendidos4.index.to_list(), menosvendidos4.to_list(), color = 'lightcoral', width = 0.1,zorder = 3)
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.grid(axis='y',zorder = 0)

plt.tight_layout()
plt.show()
```

```python
plt.bar(['Tienda 1','Tienda 2','Tienda 3','Tienda 4'],[promedioenvio1,promedioenvio2,promedioenvio3,promedioenvio4], color = ['cornflowerblue','lightcoral','mediumpurple','mediumseagreen'])
plt.xlabel('Tiendas')
plt.ylabel('Envío Promedio')
plt.show()
```

```python
ventasporlugar1 = tienda1.loc[:, ['lat','lon']]
ventasporlugar2 = tienda2.loc[:, ['lat','lon']]
ventasporlugar3 = tienda3.loc[:, ['lat','lon']]
ventasporlugar4 = tienda4.loc[:, ['lat','lon']]
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10, 12))
plt.subplot(4,1,1)

sc1 = ax1.scatter(ventasporlugar1['lon'], ventasporlugar1['lat'], alpha = 0.1)
ax1.set_title("Tienda 1")
ax1.set_ylabel("Latitud")

sc2 = ax2.scatter(ventasporlugar2['lon'], ventasporlugar2['lat'],  alpha = 0.1)
ax2.set_title("Tienda 2")
ax2.set_ylabel("Latitud")

sc3 = ax3.scatter(ventasporlugar3['lon'], ventasporlugar3['lat'],  alpha = 0.1)
ax3.set_title("Tienda 3")
ax3.set_ylabel("Latitud")

sc4 = ax4.scatter(ventasporlugar4['lon'], ventasporlugar4['lat'],  alpha = 0.1)
ax4.set_title("Tienda 4")
ax4.set_ylabel("Latitud")
ax4.set_xlabel("Longitud")


plt.show()
```

```python
ventasporlugar1 = tienda1.loc[:, ['lat','lon','Precio']]
ventasporlugar2 = tienda2.loc[:, ['lat','lon','Precio']]
ventasporlugar3 = tienda3.loc[:, ['lat','lon','Precio']]
ventasporlugar4 = tienda4.loc[:, ['lat','lon','Precio']]
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10, 12))
plt.subplot(4,1,1)

sc1 = ax1.scatter(ventasporlugar1['lon'], ventasporlugar1['lat'], c=ventasporlugar1['Precio'], cmap='Blues')
ax1.set_title("Tienda 1")
ax1.set_ylabel("Latitud")

sc2 = ax2.scatter(ventasporlugar2['lon'], ventasporlugar2['lat'], c=ventasporlugar2['Precio'], cmap='Reds')
ax2.set_title("Tienda 2")
ax2.set_ylabel("Latitud")

sc3 = ax3.scatter(ventasporlugar3['lon'], ventasporlugar3['lat'], c=ventasporlugar3['Precio'], cmap='Purples')
ax3.set_title("Tienda 3")
ax3.set_ylabel("Latitud")

sc4 = ax4.scatter(ventasporlugar4['lon'], ventasporlugar4['lat'], c=ventasporlugar4['Precio'], cmap='Greens')
ax4.set_title("Tienda 4")
ax4.set_ylabel("Latitud")
ax4.set_xlabel("Longitud")


plt.show()
```

Por lo que podemos observar, la tienda número 4 es la que peor rendimiento tiene. Aunque no parezca tener un rendimiento subóptimo en ninguna de las categorías en función de la cantidad de ventas (excepto en la de electrodomésticos), ni parece haber una cantidad de ventas general exageradamente inferior para los productos; la combinación de sus bajos ingresos con su baja calificación promedio parecen indicarnos que esta es la que peor performa entre las 4. Por otro lado, tiene un costo de envío menor al resto, y en el análisis geográfico la cantidad de ventas parece ser la misma en todas las tiendas, asi que no pude sacar conclusiones de esto. Podría deberse a que hay envíos de mayor precio a sectores lejanos en las dos primeras tiendas, que son las que tienen mayor costo de envío en promedio. Si estos productos de mayor precio fuesen de tamaño grande, o frágiles, podría causar este aumento en el precio del envío.
